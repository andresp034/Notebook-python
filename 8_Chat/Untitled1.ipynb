```python
# Función de apoyo para almacenar el estandar temporal en el archivo JSON definitivo
def trasladarAEstandar(estandarFileName, dataAux):
    with open(estandarFileName, "w", encoding = 'utf-8') as fileEstandar:
    
        json.dump(dataAux,fileEstandar,ensure_ascii=False,indent=4)
    
    fileEstandar.close()
```

```python

with open(estandarFileName, "r", encoding = 'utf-8') as fileEstandar:     # Json principal con las cuentas estandarizadas y sus patrones

    with open(jsonFileName, "w", encoding = 'utf-8') as fileAux:          # Json auxiliar para la carga de nuevos balances
        
        
        
        dataJson = json.load(fileEstandar)    # Data estandar
        patrones = []
        cont = 0
        data = {"cuentas": []}
        
        dataAux = {"cuentas": []}             # Data temporal
        
        for data in (dataJson["cuentas"]):
            
            #print(dataJson["cuentas"][cont]["patrones"])
            patrones = compararElementos(csvFileName, dataJson["cuentas"][cont]["tag"], dataJson["cuentas"][cont]["patrones"])
            
            jsonString = '{ \n"tag": "' + data["tag"] + '",\n "patrones": [' + patrones + '],\n "respuestas": ["' + data["tag"] + '"] \n}'
            #print(jsonString)
            jsonString = json.loads(jsonString)
            dataAux["cuentas"].append(jsonString)
            cont = cont + 1
        json.dump(dataAux,fileAux,ensure_ascii=False,indent=4)        # Se pasan los elementos a formato JSON
        fileEstandar.close()
        trasladarAEstandar(estandarFileName, dataAux)                 # Funcion para almacenar el documento final con el estandar
        
    fileAux.close()

    
```
